**`Matplotlib`** adalah paket yang dapat digunakan untuk membuat grafik 2d, seperti histogram, scatter plot, dan grafik-grafik lainnya pada python. **`Matplotlib`** dikembangkan sedemikian rupa agar memiliki sintaks yang mudah dipahami, dan memiliki kimiripan dengan sintax pada `MatLab`.
- Untuk menginstall, pada console: `pip install matplotlib`
- Untuk menggunakan `matplotlib` pada projek python gunakan `import matplotlib.pyplot as plt`

```python
import matplotlib.pyplot as plt
```

```python
'''inisasi data random'''
import numpy as np
X = np.random.random(size=(10000,1))
y = np.random.normal(size=(10000,1))
```

# 1 Histogram <a class="anchor" id="1"></a>
## 1.1. Inisiasi grafik <a class="anchor" id="1.1"></a>
Inisiasi grafik dengan menggunakan `matplotlib` dilakuakn dengan plt.figure(*args) yang diikuti dengan perintah pembuatan grafik. Dinyatakan dalam:


plt.figure(args)
plt.graphic_command1()
plt.graphic_command2()

argument yang dapat dimasukkan kedalam perintah `plt.figure()` antra lain:
- num = int,
- figsize = (i,j)
- dpi = (int) (resolusi gambar)
- frameon = True,

## 1.2. Membuat Histogram <a class="anchor" id="1.2"></a>

**Linspace** adalah -> Membuat array dengan nilai dalam interval

**Bins** adalah jumlah rentang data / bars. Kenapa semi wajib? Karena jika tidak disetting, pyplot akan secara otomatis mensetting nilainya sebesar 10
- np.linspace(1, 10, 2) # linespace(start, stop, number(berapa banyak))
- bins=10(default) or combine with np.linspace

**rwidth** berguna untuk memberikan space jarak antara bar 

```python
plt.hist(X, bins= np.linspace(-3, 3, 31), color='red', alpha=0.3, label='nilai X')
plt.hist(y, bins= np.linspace(-3, 3, 31), color='blue', alpha=0.3, label='nilai Y')

plt.title('ini judul')
plt.xlabel('x label')
plt.ylabel('y label')

plt.legend()
```

```python
plt.hist(X, bins= 10, color='red', alpha=0.3, label='nilai X',rwidth=1.2)
```

## 1.3 Memberi Aksesoris pada Plot <a class="anchor" id="1.3"></a>

```python
spaces = np.linspace(-1,1,100)
plt.hist(y, bins=spaces);

plt.title('Contoh Histogram')
plt.xlabel('x lable')
plt.ylabel('contoh y lable')
```

## 1.4 Mengatur Axis pada Matplotlib <a class="anchor" id="1.4"></a>

### Mengatur nilai maksimum dan minimum pada Plot
Pengaturan nilai maksimum dan minimum dapat dilakukan dengan perintah:

`plt.xlim(min_boun, max_bound)` atau `plt.ylim(min_boun, max_bound)`

```python
plt.hist(y)
plt.ylim(0,2000)
```

### Mengatur Axis text pada Plot
Text pada plot di matplotlib dapat diubah dengan menggunakan perintah:

`plt.xticks(location, texts)` atau ``plt.yticks(location, texts)``

```python
plt.hist(y)
plt.ylim(0,2000)
plt.xticks([0,2,2.5,4], 
           ['satu', 'dua', 'tiga', 'empat'], 
           rotation=90);
```

### Mengatur Skala pada Axis dan membalikkannya
Skala pada axis dapat diubah dari linear menjadi Logaritmik dengan fungsi: `plt.gca().set_xscale('log')` atau `plt.gca().set_yscale('log')`

Untuk membalikkan axis dapat menggunakan `plt.gca().invert_xaxis()` atau `plt.gca().invert_yaxis()`



```python
plt.hist(y)
plt.ylim(0.2,2000)
plt.xticks([0,2,2.5,4], 
           ['satu', 'dua', 'tiga', 'empat'], 
           rotation=90);


plt.gca().set_yscale('log')
plt.gca().invert_yaxis()
```

# 2 Lineplot <a class="anchor" id="2"></a>
Lineplot adalah diagram garis. Data yang diperlukan adalah data X dan Y dalam bentuk array atau dalam bentuk DataFrame. Note: X dan Y harus memiliki panjang yang sama.<br>
Perintah yang digunakan untuk membuat lineplot adalah:
- `plt.plot(args)`
    - X: data dimensi X
    - y: data dimensi y
    - format garis dalam bentuk teks. Contoh format garis: 
        - '-g' = garis tegas berwarna hijau
        - '--r' = garis putus-putus berwarna merah


```python
plt.plot(X,y)
```

```python
'''creating dummy data'''
X1 = [1,3,2,9]
y1 = [9,7,4,1]

plt.plot(X1, y1, '--<r')
```

```python
plt.plot(X1,y1,linewidth=2, marker ='<', markerfacecolor='yellow', linestyle='--', color='red')
```

# 3 Scatter Plot <a class="anchor" id="3"></a>
Scatter plot adalah diagram titik. Data yang diperlukan adalah data X dan Y dalam bentuk array atau dalam bentuk DataFrame. Note: X dan Y harus memiliki panjang yang sama.<br>
Perintah yang digunakan untuk membuat lineplot adalah:

- `plt.scatter(args)`
    - X: data dimensi X
    - y: data dimensi y
    - c: warna (bisa dalam bentuk array berukuran yang sama dengan X)
    - s: ukuran marker (bisa dalam bentuk array berukuran yang sama dengan X)
    - alpha: transparansi dari data (bisa dalam bentuk array berukuran yang sama dengan X)


```python
X = np.random.random(size=(10000,1))
y = np.random.normal(size=(10000,1))
```

```python
plt.scatter(X,y,c='red', alpha=0.1, s=X*25)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X,y,s=X*50,alpha=0.1,c='g')
```

## 5.1 Menggunakan Subplot <a class="anchor" id="5.1"></a>
Subplot adalah perintah untuk membagi figure kedalam panel dengan ukuran (i * j). dengan sintax:
- `plt.subplot(i,j,n)`
- i = banyak kolom
- j = banyak baris
- n = urutan


```python
'''creating dummy data'''
X1 = [1,6,7,9,12,14]
y1 = [-2,2,5,5,7,9]

X2 = [2,1,3,4,7]
y2 = [9,7,3,2,-1]
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.scatter(X1,y1)
plt.title('grafik 1')

plt.subplot(1,2,2)
plt.plot(X1,y1,'r')
plt.title('grafik 2')
```

# 6 Seaborn
Seaborn adalah library visualisasi yang drancang untuk mempermudah analisa EDA dengan data tabel. Kebelihan dari seaborn adalah:
- Mempermudah plotting untuk data cetegorical
- Mempermudah plotting pada data dengan kolom yang banyak

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('file/automobile_data.csv', delimiter=';')
```

```python
sns.scatterplot('peak-rpm', 'num-of-cylinders', 
                data=df,
                hue = 'drive-wheels')
```

```python
sns.boxplot(x='drive-wheels', y = 'highway-mpg' , data=df)
```

```python
sns.violinplot(x='drive-wheels', y = 'highway-mpg' , data=df)
```

```python
sns.pairplot(df,hue='drive-wheels')
```

```python
sns.lmplot('width', 'height',data=df,)
```

```python
sns.heatmap(df.isna())
```

```python
sns.heatmap(df.corr())
```

```python
a = plt.hist(df['height'] )
```

```python
sns.kdeplot(df['height'])
plt.plot(a[1][1:],a[0]/np.sum(a[0]))
plt.scatter(df['height'],[0]*len(df['height']),alpha=0.2)
```
